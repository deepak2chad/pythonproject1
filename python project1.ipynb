```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('studentscore.csv')
print(data.head(5))
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# drop unnamed coloumn
```

```python
df=data.drop("Unnamed: 0",axis=1)
print(df.head(10))
```

```python
#GENDER DISTRIBUTION
```

```python
plt.figure(figsize=(4,5))
x1=sns.countplot(data=df,x='Gender')
x1.bar_label(x1.containers[0])
plt.show()
```

```python
#conclusion-the number of female is more than male
```

```python
gb=df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
gb
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship between parent education and students'scores")
plt.show()
```

```python
#CONCLUSION-From above we conclude that parents education have huge impact on thier scores
```

```python
gb1=df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
gb1
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between parent's marital status and students score")
plt.show()
```

```python
#CONCLUSION-we conclude that parent marital status have negligible impact on student socres. 
```

```python
groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()

list=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
print(list)
X=['group A','group B','group C','group D','group E']
plt.pie(list,labels=X,autopct='%1.2f%%')
plt.title('Distribution of Ethnic GROUPS')
plt.show()
```
